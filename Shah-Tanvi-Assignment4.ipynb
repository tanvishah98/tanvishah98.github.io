```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import ipywidgets 
#imported the necessary libraries
```

```python
xnames = ['date','report_date','longitude','latitude','duration_seconds','state','country','shape']
ynames = ['report_date','longitude','latitude','duration_seconds']
color = ['Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds',
            'YlOrBr', 'YlOrRd', 'OrRd', 'PuRd', 'RdPu', 'BuPu',
            'GnBu', 'PuBu', 'YlGnBu', 'PuBuGn', 'BuGn', 'YlGn']

@ipywidgets.interact(style=plt.style.available, x=xnames,y=ynames, cmap=color)
# @ipy.interact()
def make_plot(style,x,y,cmap):
    with plt.style.context(style):
        fig, ax = plt.subplots(1,1,figsize=(10,10))
        plt.scatter(ufoDS[x],ufoDS[y], c=np.log10(ufoDS['duration_seconds']), cmap = cmap)
        if x == 'state' or x == 'shape':
            plt.xticks(rotation=90)
        ax.set_ylabel(y)
        ax.set_xlabel(x)


    plt.show()
```

For the first part, we had to allow the user to choose the x and y fields on a scatter plot. So I created two variables namely 'xnames' and 'ynames' in which the fields are specified to plot on a scatter plot. The colours for the color map are mentioned as well. By using ipywidgets.interact, there were 4 dropdowns for style, x field, y field and colormap. 'Duration_seconds' is used as the color scaling for the plots. After plotting, the city and shape on the x axis are not easily readible which is why the names on the x axis are rotated vertically. 

Basically the user is allowed to choose the style of the plot, x, y, and the color map and accordingly the plot is created. 
I did not use 'city' as one of the field because there is a huge number of different cities and while plotting them on the x axis none of the city names were readable even after using a small sample amount of data. I did not use duration as one of the fields because the data is not in one format. One row value is in mins and the other is in hours. I also did not use comments because they are just basic comments in the form of small paragraphs.

For the y axis, only numerical variables are used. Along with that, report_date is one of the fields for y axis and date is thee field in x axis. Having both these feilds on both the axis does not make sense as there will only be a y=x axis plotted. 
One of the challenges I went through in this question was to assign the color maps and also to understand what columns should be available for the user to choose from for the x and y field.

```python
irange = ipywidgets.IntSlider(min=1, max=100, step=1) 
xnames = ['date','report_date','longitude','latitude','duration_seconds','state','country','shape']

@ipywidgets.interact(style=plt.style.available, x=xnames,bins=irange)
# @ipy.interact()
def make_plot(style,x,bins):
    if(x=='country' or x=='shape'or x=='state'):
        with plt.style.context(style):
            fig, ax = plt.subplots(1,1,figsize=(10,10))
            plt.bar(ufoDS[x].value_counts().index, height = ufoDS[x].value_counts().values)
            if x == 'state' or x == 'shape':
                plt.xticks(rotation=90)
    else:
        with plt.style.context(style):
            fig, ax = plt.subplots(1,1,figsize=(10,10))
            plt.hist(ufoDS[x], bins=bins)
            if x == 'state' or x == 'shape':
                plt.xticks(rotation=90)
        
    ax.set_xlabel(x)


    plt.show()
```

For the second part, we had to display aggregate values of a column and the user should be able to selext the field as well as the number of bins for numerical variables. So along with the column names mentioned for the x field, i created an integer slider as well to select the number of bins. Similar to the first question, ipywidgets.interact is used which created 3 dropdowns for style, x field and bins if the variable is numerical. First we determined which columns can be used to make bin plotsand then we checked if the selected column is a categorical variable or numerical, and if it is a categorical variable then a bar plot is plotted. If it is a numerical variable, a histogram is plotted and the user can select the number of bins using the slider. State, Country, Shape are categorical variables and the others come under numerical variable.

For the number of bins, I specifically used the Integer slider as it was very interesting to me and wanted to incorporate such widgets in my questions after learning about them more in class. 

City, comments and duration are not used as the fields in the x axis for the same reason as the first part.
